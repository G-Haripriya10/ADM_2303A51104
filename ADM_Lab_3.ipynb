```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([3,8,1,7])
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A","B","C","D"])
y= np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
myexplode = [0.2,0,0,0]
plt.pie(y,labels = mylabels, explode=myexplode, shadow=True)
plt.show()
```
